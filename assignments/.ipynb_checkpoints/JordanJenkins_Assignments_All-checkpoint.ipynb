### Feature Engineering
Some columns are not useful for analysis as they currently stand. I will be converting the 52 week range into two columns representing the min and max 52 week value, the daily price range into daily min and max values, and the price change into a percent change and absolute change value.

```python
# Split out 52 week price range into min and max value
new = df['52 week range'].str.split(' - ', n=1, expand=True)
df['min_52'] = new[0]
df['max_52'] = new[1]
df.drop(columns='52 week range', inplace=True)
```

```python
# Split daily range into min and max value
new = df['price range'].str.split(' - ', n=1, expand=True)
df['min_daily'] = new[0]
df['max_daily'] = new[1]
df.drop(columns='price range', inplace=True)
```

### Non-categorical to categorical
It may be useful to look at volume as it relates to the day's average volume. If the volume is much higher than average, there could be a lot of volatility in the price. I will split the column into the following:
* Higher than Average (greater than the 75th percentile for the individual stock)
* Average (between the 25th and 75th percentile)
* Below Average (below the 25th percentile)

```python
# I initially wanted to use pandas.qcut, but there were several companies with volume values that did not change.
# And I kept ketting the duplicate edge error, when dropping duplicates, I would get errors. This function
# defines the percentile edges and then applies this rank to values.

def pct_rank_qcut(series, n):
    edges = pd.Series([float(i) / n for i in range(n + 1)])
    f = lambda x: (edges >= x).argmax()
    return series.rank(pct=1).apply(f)
```

```python
# Group the data by company name and apply the custom qcut function
df['vol_cat'] = df.groupby('company_name')['volume'].transform(pct_rank_qcut,4)
df[['company_name', 'volume', 'vol_cat']]
```

```python
# Change the resulting `vol_cat` values to those defined in the header of this assignment.
df['vol_cat'] = np.where(
     df['vol_cat'].between(0, 1), 
    'Below_Avg', 
     np.where(
        df['vol_cat'].between(1, 3), 'Avg', 'Above_Avg'
     )
)
df.sample(20)
```

```python
df['vol_cat'] = df['vol_cat'].astype('category')
df.info()
```

### Use Regular Expressions to Clean a Column

```python
df['market cap'].str[-1].unique()
```

The Market Cap column is a string that ends in 'M', 'B', 'T', or is '0.00'. I will use regular expressions to extract the numeric value and convert it to a common scale (Billion).

```python
import re
```

```python
'''
\d+ digits before optional decimal
.? optional decimal(optional due to the ? quantifier)
\d* optional digits after decimal
'''
pattern = r'(\d+\.?\d*)'

df['mk_val'] = df['market cap'].str.extract(pattern).astype(float)
```

```python
pattern = r'([a-zA-Z]+)'
df['mk_multiplier'] = df['market cap'].str.extract(pattern)
```

```python
def unit_conv(row):
    val = row['mk_val']
    unit = row['mk_multiplier']
    
    if unit == 'T':
        return val * 1000 # if unit is Trillion, multiply by 1000 to get billions
    elif unit == 'M':
        return val / 1000 # if unit is Million, divide by 1000 to get billions
    else:
        return val
```

```python
df['mk_cap_cleaned_billions'] = df.apply(unit_conv, axis=1)
```

```python
df[['mk_val', 'mk_multiplier', 'mk_cap_cleaned_billions']].sample(20)
```
